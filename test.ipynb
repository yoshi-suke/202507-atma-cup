```python
from sklearn.model_selection import StratifiedKFold
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix

# X = train_merge.iloc[:, 8:]
X = train_merge.drop(columns=["社員番号", "is_test", "target"])
# X = train_merge.drop(columns=["社員番号", "target"]) # No Categの場合
y = train_merge["target"]


# クロスバリデーション：層化5分割
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

for fold, (train_idx, test_idx) in enumerate(skf.split(X, y)):
    print("fold: ",)
    X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
    y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]
    
    clf = RandomForestClassifier(
        random_state=42, 
        max_depth=3,
        min_samples_leaf=3,
        min_samples_split=5,
        n_estimators=500,
        # max_samples=1000,
        # max_features="log2",
        class_weight='balanced')
    clf.fit(X_train, y_train)
    y_pred_train = clf.predict(X_train)
    y_pred = clf.predict(X_test)
    
    # 混同行列
    cm_train = confusion_matrix(y_train, y_pred_train)
    cm = confusion_matrix(y_test, y_pred)
    
    print(f'Fold {fold}')
    print('Confusion Matrix (Train):')
    print(cm_train)
    print('Confusion Matrix (Test):')
    print(cm)
    print()
```

```python
from statsmodels.stats.proportion import proportions_ztest
def ztest(df, col, target_col):
    a_pos = df[df[col]>=1][target_col].sum()
    a_sample = len(df[df[col]>=1])
    b_pos = df[df[col]==0][target_col].sum()
    b_sample = len(df[df[col]==0])
    # 成功数
    successes = [a_pos, b_pos]
    # 試行数
    samples = [a_sample, b_sample]

    # 検定（帰無仮説：比率に差がない）
    stat, pval = proportions_ztest(count=successes, nobs=samples)

    # Cohen's h 効果量
    cohen_h = 2 * (np.arcsin(np.sqrt(a_pos/a_sample)) - np.arcsin(np.sqrt(b_pos/b_sample)))

    print(f'z値: {stat:.3f}')
    print(f'p値: {pval:.3f}')
    print(f"Cohen's h: {cohen_h:.3f}")

    if abs(cohen_h) >= 0.2:
        print("効果量あり")
        return True
    else:
        print("効果量なし")
        return False
```

## モデル作成（Feature Engineering後）

### RandomForest

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix

# X = train_merge.iloc[:, 8:]
X = train_merge[features_cols]
y = train_merge["target"]


# クロスバリデーション：層化5分割
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

for fold, (train_idx, test_idx) in enumerate(skf.split(X, y)):
    print("fold: ",)
    X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
    y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]
    
    clf = RandomForestClassifier(
        random_state=42, 
        # max_depth=3,
        # min_samples_leaf=3,
        # min_samples_split=5,
        n_estimators=500,
        # max_samples=1000,
        # max_features="log2",
        class_weight='balanced')
    clf.fit(X_train, y_train)
    y_pred_train = clf.predict(X_train)
    y_pred = clf.predict(X_test)
    
    # 混同行列
    cm_train = confusion_matrix(y_train, y_pred_train)
    cm = confusion_matrix(y_test, y_pred)
    
    print(f'Fold {fold}')
    print('Confusion Matrix (Train):')
    print(cm_train)
    print('Confusion Matrix (Test):')
    print(cm)
    print()
```

```python
### 変数重要度の可視化 ###
features =    X.columns                     # 説明変数名称を取得
importances = clf.feature_importances_   # 変数重要度の取得
importance_df = pd.DataFrame({
    'feature': features,
    'importance': importances
})
importance_df = importance_df.sort_values('importance', ascending=False).reset_index(drop=True)
importance_df['importance'] = importance_df['importance'].apply(lambda x: '{:.6f}'.format(x))
importance_df
```

### LogisticRegression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score, roc_curve

# X = train_merge.iloc[:, 8:]
X = train_merge[features_cols]
y = train_merge["target"]


# クロスバリデーション：層化5分割
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

for fold, (train_idx, test_idx) in enumerate(skf.split(X, y)):
    print("fold: ",)
    X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
    y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]
    
    model = LogisticRegression(
        # C=0.1,
        # penalty="elasticnet",
        # l1_ratio=0.1,
        # solver="saga",
        max_iter=1000, # Default(100)で収束しなかったので大きい値に設定
        class_weight="balanced"
        ) 
    model.fit(X_train, y_train)
    
    # (1) 予測確率を取得
    y_pred_proba_train = model.predict_proba(X_train)[:,1]  # positiveクラス（1）の確率
    y_pred_proba = model.predict_proba(X_test)[:,1]  # positiveクラス（1）の確率

    # (2) ROC-AUC算出
    auc_train = roc_auc_score(y_train, y_pred_proba_train)
    auc = roc_auc_score(y_test, y_pred_proba)

    print(f'Fold {fold}')
    print('ROC-AUC (Train):', auc_train)
    print('ROC-AUC (Test):', auc)
    print()
```

```python
plt.hist(y_pred_proba, bins=50, range=(0,1), color='skyblue', edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of array values (min=0, max=1)')
plt.xlim(0, 1) 
```

### XGBoost

```python
import xgboost as xgb
from sklearn.metrics import roc_auc_score, roc_curve

# X = train_merge.iloc[:, 8:]
X = train_merge[features_cols]
y = train_merge["target"]


# クロスバリデーション：層化5分割
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

for fold, (train_idx, test_idx) in enumerate(skf.split(X, y)):
    print("fold: ",)
    X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
    y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]
    
    # XGBoost（scikit-learn互換APIのXGBClassifierを利用）
    model = xgb.XGBClassifier(
        use_label_encoder=False,        # 警告回避用。xgboost 1.0以降では非推奨
        eval_metric='logloss',         # 二値分類の場合
        random_state=42,
        max_depth=3,
        min_child_weight=5,
        learning_rate=0.01,
        # gamma=0.1,
        scale_pos_weight=24,
        # colsample_bytree=0.8,
        n_estimators=1000
    )
    model.fit(X_train, y_train)
    
    # (1) 予測確率を取得
    y_pred_proba_train = model.predict_proba(X_train)[:,1]  # positiveクラス（1）の確率
    y_pred_proba = model.predict_proba(X_test)[:,1]  # positiveクラス（1）の確率

    # (2) ROC-AUC算出
    auc_train = roc_auc_score(y_train, y_pred_proba_train)
    auc = roc_auc_score(y_test, y_pred_proba)

    print(f'Fold {fold}')
    print('ROC-AUC (Train):', auc_train)
    print('ROC-AUC (Test):', auc)
    print()
```
